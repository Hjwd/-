```python
import numpy as np
import matplotlib.pyplot as plt
import random

plt.rcParams["figure.figsize"] = (5, 5)
lr=0.1 #步长
precision=1e-6 #精度(视为收敛的梯度值)

def function(x): # 求函数值
    return (x-1)**2

def derivatives(x): # 求导数
    return 2*(x-1)

def gradientDescent(x,epoch):# 梯度下降
    if derivatives(x)>precision:
        epoch=epoch+1;
        plt.scatter(x, function(x), marker='x', color='red') #描点标记
        new_x=x-lr*derivatives(x) # 计算新的X点
        print("迭代次数：",epoch,"梯度：",derivatives(x)," f(x):",function(x)) # 输出每次迭代的内容
        return gradientDescent(new_x,epoch) # 迭代
        
x_point=np.linspace(-10,10)
plt.plot(x_point, function(x_point)) # 画出图像
gradientDescent(8,0)
plt.show()
    
    
```

```python
# 房价预测
import numpy as np
import random
import matplotlib.pyplot as plt

# 学习率
alpha=0.01

# 迭代数
epoch=10000

# 函数模型
def model(a,b,c,x1,x2): 
    return a+b*x1+c*x2

# 梯度
def derivatives_all(a,b,c,x1,x2,y):
    a_=np.mean(model(a,b,c,x1,x2)-y)
    b_=np.mean((model(a,b,c,x1,x2)-y)*x1)
    c_=np.mean((model(a,b,c,x1,x2)-y)*x2)
    return a_,b_,c_

# 损失函数
def lossFunction(prediction,y):
    return np.mean((prediction-y)**2)

# 批量梯度下降
def gradientDescent_random(a,b,c,x1,x2,y):
    loss_random=[]
    for i in range(epoch):
        loss_random.append(lossFunction(model(a,b,c,x1,x2),y))
        a_,b_,c_=derivatives_all(a,b,c,x1,x2,y)
        a=a-a_*alpha
        b=b-b_*alpha
        c=c-c_*alpha
    print("final_a:",a," final_b:",b," final_c",c," final_loss:",loss_random[-1])
    return loss_random

# 随机梯度下降
def gradientDescent_part(a,b,c,x1,x2,y):
    loss_part=[]
    for i in range(10):
        random_array = np.random.randint(0, 1000, size=100)
        selected_x1 = x1[random_array]
        selected_x2 = x2[random_array]
        selected_y = y[random_array]
        for j in range(int(epoch/10)):
            a_,b_,c_=derivatives_all(a,b,c,selected_x1,selected_x2,selected_y)
            a=a-a_*alpha
            b=b-b_*alpha
            c=c-c_*alpha
            loss_part.append(lossFunction(model(a,b,c,selected_x1,selected_x2),selected_y))
    print("final_a:",a," final_b:",b," final_c",c," final_loss:",loss_part[-1])
    return loss_part

# 生成一些示例数据
np.random.seed(0)
x1 = np.random.rand(1000)
x2 = np.random.rand(1000)

#预期的函数附近生成数据y
y = 2 * x1 + 2 * x2 + 1 +0.1* np.random.randn(1000)

# 随机生成超参数
a = np.random.randn(1)
b = np.random.randn(1)
c = np.random.randn(1)
loss1=gradientDescent_random(a,b,c,x1,x2,y)
loss2=gradientDescent_part(a,b,c,x1,x2,y)

plt.subplot(1,2,1)
plt.plot(range(len(loss1)),loss1,label="loss_part")
plt.show()
plt.subplot(1,2,1)
plt.plot(range(len(loss2)),loss2,label="loss_random")
plt.show()


```
